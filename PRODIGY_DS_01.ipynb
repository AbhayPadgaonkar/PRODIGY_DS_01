# Importing necessary Python libraries

```python
import pandas as pd 
import seaborn as sn 
import matplotlib.pyplot as plt
```

# Data loading and filtering

```python
df = pd.read_csv('Population of India.csv')
df.head()
```

```python
df['Population'] = df['Population']/10000000
df['Male'] = df['Male']/10000000
df['Female'] = df['Female']/10000000
df = df.rename(columns={'Population' : 'Population(Crores)' , 'Male' : 'Male(Crores)','Female' : 'Female(Crores)'})
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
#Creating another data frame with the last row of df in order to filter that row out from df as we only want state values for state based analysis and total values for trends in india
df_total = df.iloc[36]
display(df_total)
```

```python
df = df.iloc[:36]
df.tail()
```

# State based population visualization

```python
plt.figure(figsize=(10, 10))
sn.barplot(x='Population(Crores)', y='State/UT', data = df).set_title('State based population')
```


# State based Sex Ratio visualization

```python
plt.figure(figsize=(10, 10))
sn.barplot(x ='Sex ratio', y ='State/UT', data = df).set_title('State based sex ratio')
```

# State based Male and Female population visualization 

```python
plt.figure(figsize=(10, 10))
df_melted = df.melt(id_vars="State/UT", value_vars=["Male(Crores)", "Female(Crores)"], var_name="Gender", value_name="Population")
sn.barplot(x = 'Population', y = 'State/UT', hue = 'Gender', data = df_melted, dodge = True)
```
